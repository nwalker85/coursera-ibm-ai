KNN basic outline

1. Pick a value for k
2. calculate the distance of unknown case from all cases
3. select the k-observations in the training data that are "nearest" to the unknown data point
4. predict the response of the unknown data point using the most populare response value from the k-nearest neighbors


You can find the distance between two data points by using the euclidean distance.

```python
from scipy.spatial import distance
a = (54, 190, 3)
b = (50, 200, 8)
dst = distance.euclidean(a, b)
print(dst)
```

so you would loop through each permutation of your data, find the k smallest dst

```python


tuples = list(cdf.itertuples(index=False, name=None))
print(len(tuples))
for x in range(len(tuples)):
    for y in range(len(tuples)):
        if x == y: continue
        dst = distance.euclidean(tuples[x], tuples[y])
        print(dst)


```

```python
for x in range(len(cdf.index)):
    for y in range(len(cdf.index)):
        if x == y: continue
        print(x, y)
        print(cdf.iloc[x], cdf.iloc[y])
```

you need to be careful about the size of k

too small and you overfit, 
too big and you overgeneralize
