# Recursive Partitioning

"Predictiveness" of a given independant variable
The basic idea is, group an independant variable by dependant variable

What is entropy?

```python
labels = ['B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'A', 'A', 'A', 'A', 'A']

from collections import Counter
from scipy import stats

stats.entropy(list(Counter(labels).values()), base=2)
```

```python
labels = ['B', 'B', 'B', 'B', 'B', 'B', 'A', 'A']

stats.entropy(list(Counter(labels).values()), base=2)
```

```python
labels = ['B', 'B', 'B', 'A', 'A', 'A']

stats.entropy(list(Counter(labels).values()), base=2)
```

Calculate which has the higher information gain, and you'll know that independant variable is more correllated
