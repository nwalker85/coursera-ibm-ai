Logistic Regression

Tries to predict a categorical or discreet target field instead of a continous variable.

this is contrasted from linear regression, which is used to find a value, logistic regression is for booleans

When should we use logistic regression?

When the result you are looking for is binary

If you need probability results (between 0-1)

When you need a linear decision boundary

You need to understand the impact of a feature
