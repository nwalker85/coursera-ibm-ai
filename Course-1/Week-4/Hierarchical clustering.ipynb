# TLDR

1) Create n clusters, one for each data point.

2) Compute the proximity matrix, which will be an n by n table. 

3) Do the following until you have desired clusters:
    1) Merge the two closest clusters
    2) Update the proximity matrix

## Pros and Cons to Hierarchial clustering

### Pros

doesn't require number of clusters to be specified

easy to implement

produces a dendrogram, which helps with understanding the data

### Cons

Can never undo any previous steps throughout the algorithm

generally has long runtimes

Sometimes difficult to identify the number of clusters by the dendrogram

## Hierarchical vs K-means

### k-means

much more efficient

requires the number of clusters to be specified

gives only one partition of the data based on the predefined number of clusters

potentially retuns different clusters each time it is run due to the random initialization of centroids

### hierarchical

can be slow for large datasets

does not require the number of clusters to run

gives more than one partition depending on the resolution

consistent - always generates the same clusters
