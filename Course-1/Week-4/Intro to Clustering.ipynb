# Clustering

This is the term we use for unsupervised grouping of data in a dataset.

A CLUSTER is a group of objects that a similar to each other and dissimilar to other groups.

# This is distinct from a CLASSIFICATION algorithm like SVM, decision tree, or logistical regression, in that with classification, you are doing SUPERVISED training, using LABELED data, to determine a categorical class label (predict an independant variable), while in CLUSTERING, you do UNSUPERVISED training, using UNLABELED data, to group your data based on similar attributes.

You can use this concept to find a group of similar items or users to recommend products. 

Partition-based clustering
    sphere like clusters on your graph
        ex k-Means, k-Meadia, Fuzzy c-Means
 
Hieracrchical clustering
    tree like clusters, good for small data sizes.
        ex Agglomerative, Divisive

Density-based Clustering
    arbitrary shaped clusters
        ex DBSCAN
