k-means divides the datya into non-overlapping subsets (clusters) without any cluster-internal structure

Examples within a cluster are very similar

Examples across different clusters are very different.

How do we find these groups and then build metrics to measure them?

### K means looks at the differences and tries to minimize the intra-cluster distances and maximize the inter-cluster distances.

The euclidian difference shows its head again.

Now move centroids to the mean of the datapoints within the cluster.

Now go back and do this again until the centroids no longer move.

Now, the crazy thing is, the starting points influence the end result, so you have to run this repeatedly using different starting points to get a valid result.

1) Randomly place k centroids.

2) Calculate the euclidean distance of each datapoint from each centroid. 

3) Assign each data point to its closest centroid, creating a cluster.

4) Recalculate the position of the k centroids as the mean of the data.

5) repeat until the centroids stop moving.

You have two real options for metrics here.

Compare the clusters with the ground truth if it's available, but we usually don't in the real world.

You'll mostly use the average distance between datapoints within a cluster. Also, the average of the distance of each datapoint from their centroid can be a metric. 

Determining the correct k value is hard because it depends on the shape/scale of the data. Common approach is to run the approach with multiple values of k and determine the most accurate.

In theory, you can indefinitely increase k and keep seeing improvement, so look for the elbow point. K=5 is the correct value to pick here.
