# DENSITY BASED SPATIAL CLUSTERING OF APPLICATIONS WITH NOISE

Density based clustering algorithm, best used for spacial data.

When applied to arbitrary shaped clusters, or clusters within clusters, the basic sphere shapes just don't work.

k-means doesn't account for outliers.





The basic idea is that ignores less dense areas.

r = radius
m = minimum number of data points within a radius to be considered a "cluster"

each point is either a core point, border point, or outlier point.

look at each data point and determine it's type.

a datapoint is a core point, if within r, there are m points.

a datapoint is a boarder point, if within r is less than m but contains a core point.

a datapoint is an outlier, if within r is less than m and does not contain a core point.



once all points are labeled, connect core points that are neighbors, and put them in the same cluster.



Advantages:

1) good with arbitrariy shaped clusers.

2) robust to outliers

3) very practical because you don't need to specify the clusters.

This seems like it's clearly the best one
